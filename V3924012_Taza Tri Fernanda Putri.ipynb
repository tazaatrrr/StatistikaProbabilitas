```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter distribusi normal
mean = 50
std_dev = 5
n = 100  # jumlah nilai acak yang dihasilkan

# Menghasilkan 100 nilai acak
data = np.random.normal(mean, std_dev, n)

# Menghitung mean, median, dan standar deviasi dari data yang dihasilkan
mean_data = np.mean(data)
median_data = np.median(data)
std_dev_data = np.std(data)

#membuat plot
plt.plot(data)
plt.scatter([mean], [mean_data], label="Rata-rata", color="blue")
plt.scatter([mean], [median_data], label="Median Data", color="pink")

# Menampilkan hasil
print(f"Random Data: {data}")
print(f'Mean: {mean_data}')
print(f'Median: {median_data}')
print(f'Standar Deviasi: {std_dev_data}')

#menambahkan judul,label data
plt.title('Data  distribusi normal')
plt.xlabel('indeks')
plt.ylabel('nilai')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat DataFrame
data = {
    'x': [5, 10, 15, 20, 25],
    'y': [10, 20, 20, 30, 30]
}
df = pd.DataFrame(data)

# Menghitung mean, median, dan modus
mean_x = df['x'].mean()
median_x = df['x'].median()
modus_x = df['x'].mode()[0]

mean_y = df['y'].mean()
median_y = df['y'].median()
modus_y = df['y'].mode()[0]

# Menampilkan hasil
print(f"Kolom x: Mean = {mean_x}, Median = {median_x}, Modus = {modus_x}")
print(f"Kolom y: Mean = {mean_y}, Median = {median_y}, Modus = {modus_y}")

plt.step(df['x'], df['y'], where='post', color='green', label='Data Points')
```
